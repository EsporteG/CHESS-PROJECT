```python
import pandas as pd
```

```python
df = pd.read_excel('raw_chess_players_data\lpsupi_archive.xlsx')
df = df.dropna(subset=['url'])
df = list(df['pgn'])
df = list(map(lambda value: value.split('\n\n')[0].split('\n'), df))
df_f = []

for png in df:
    df = list(map(lambda value: value.strip("[").strip("]").strip('\s').rstrip('"').replace(" ","").split('"'), png))
    df_f.append(df)
    i = 1
df_final = pd.DataFrame()

for pgn_pro in df_f:
    df = pd.DataFrame(pgn_pro)
    df = df.transpose()
    df = df.rename(columns=df.iloc[0]).drop(df.index[0])
    df['match id'] = i
    i=i+1
    df_final = pd.concat([df_final,df])
df_final.head()

```
